```python
from dataclasses import dataclass, asdict, field
import json
from typing import Dict, List


from enb import Classifier


@dataclass
class DataRow:
    id: str
    label: str
    text: str


@dataclass
class Dataset:
    train: bool = True
    rows: List[DataRow] = field(default_factory=list)


def create_dataset():
    with open("data/fake_newsgroup.json") as f:
        data = json.load(f)

    train = Dataset(train=True, rows=[DataRow(**r) for r in data['train']])
    test = Dataset(train=True, rows=[DataRow(**r) for r in data['test']])

    return train, test

create_dataset()
```
